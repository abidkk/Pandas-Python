# Group_by

```python
import pandas as pd
```

```python
d1 = pd.read_csv('../Datasets/students_set..csv')
d1
```

# Group BY

     -  Grouped by Gender

```python
dg = d1.groupby("Gender")
dg

for i in dg:
    print(i)
    print(' \n __ \n')
```

        - Grouped by City

```python
dgc = d1.groupby("City")
dgc

for i in dgc:
    print(i)
    print(' \n __ \n')
```

# Get Group
 - to get specific group

```python
dg.get_group("Female")
```
