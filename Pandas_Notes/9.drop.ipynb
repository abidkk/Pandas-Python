# Drop  Missing Values
- Dronpna()


```python
import pandas as pd
```

- Reading CSV
    - Note : missing value are represented NaN

```python
f1 = pd.read_csv('../Datasets/pupils.csv')
f1
```

# 1. Dropna : Drop all rows having any missing value

```python
f1.dropna()
```

# dropna()
- how("___")
    - any : drop row having any NaN value
    - all : drop wor having all values NAN

- subset(['column_Name'])
    - remove row having NaN value of specific column

- inplace:
    - remove rows having any NaN value and reutrn new dataset

```python
f1.dropna(how='any')
```

```python
f1.dropna(how='all')
```

```python
f1.dropna(subset=["Course"])
```

```python
f1.dropna(inplace=True)
f1
```

# 3. Dropna : on axis-1 remove all columns having missing values

```python
f1.dropna(axis=1)
```

# FILLNA 

# 1.  Fill NaN with a constant value

```python
f1.fillna(0) # all misisng values filled with 0
```

```python
f1.fillna("x") # all missing values filled will with x
```

# Forward fill NaN values (propagates previous value in column)

```python
f1.fillna(method='ffill')
```
