# Merge & Concat

```python
import pandas as pd 
```

```python
d1 = pd.DataFrame({'A':[10,20,30,40,50], 'B':[1,2,3,4,5], 'C':['a1', 'b2', 'c3', 'd4', 'e5']})
d2 = pd.DataFrame({'A':[10,20,30,40,50], 'B':[1,2,3,4,5], 'C':['a1', 'b2', 'c3', 'd4', 'e5']}) 
```

```python
pd.merge(d1,d2)
```

# Merge stoped 

```python
d1 = pd.DataFrame({'A':[10,20,30,40,50], 'B':[1,2,4,6,8]})
d2 = pd.DataFrame({'A':[10,20,30,40,60], 'C':[1,2,4,4,5]})

pd.merge(d1,d2)
```

- How Parameter to continue 

```python
d1 = pd.DataFrame({'A':[10,20,30,40,50], 'B':[1,2,4,6,8]})
d2 = pd.DataFrame({'A':[10,20,30,40,60], 'C':[1,2,4,4,5]})


pd.merge(d1,d2, how='left')
```

- Outer

```python
pd.merge(d1,d2, how='outer')
```

# Indicator

```python
pd.merge(d1,d2, how='outer', indicator=True)
```

# Left & Right Index

```python
pd.merge(d1,d2, left_index=True, right_index=True)
```

```python
pd.concat([d1,d2]) 
```

```python
pd.concat([d1,d2], axis=1) 
```

```python
d1 = pd.DataFrame({'A':[10,20,30,40,50], 'B':[1,2,3,4,5]})
d2 = pd.DataFrame({'A':[10,20,30], 'B':[1,2,3]}) 

pd.concat([d1,d2], axis=1) 
```
