# Replace Data in Dataset

```python
import pandas as pd
```

```python
d1 = pd.read_csv('../Datasets/pupils.csv')
d1
```

    - Replace Java to Typescript

```python
d1.replace(to_replace="Java", value="Typescript")
```

    - Replace Karachi and Lahore to Islamabad

```python
d1.replace(to_replace=["Karachi","Lahore"], value="Islamabad")
```

# Froward & Backword Replace
- method("___")
    - ffill
    - bfill

```python
d1.replace('Java', method='ffill')
```

```python
d1.replace('Java', method='bfill')
```

```python
d1.interpolate()
```

```python
d1.interpolate()
```

```python
d1.interpolate(method='index')
```

```python
d1.interpolate(limit=2)
```

```python
d1.interpolate(method='index', limit_direction="forward")
```
