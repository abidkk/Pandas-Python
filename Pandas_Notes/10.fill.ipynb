# Fill Missing Vakues
- fillna()

```python
import pandas as pd 
```

```python
d1 = pd.read_csv('../Datasets/pupils.csv')
d1
```

# Filling all values with constant 0

```python
d1.fillna(0)
```

# Filling all values with constant 'X'

```python
d1.fillna('x')
```

# Fill NaN values of specific column

```python
d1.fillna({'City':"Karachi"})
```

```python
d1.fillna({'City':"Karachi", 'Course':"Python"})
```

# Forward & Backward Filling
- method('___')
    - ffill
    - bfill
    

```python
d1.fillna(method='ffill')
```

```python
d1.fillna(method='bfill')
```
